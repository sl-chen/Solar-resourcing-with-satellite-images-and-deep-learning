```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Layer, Input, Conv2D, MaxPool2D, concatenate, Flatten, Dropout, BatchNormalization
from tensorflow.keras.utils import plot_model
from tensorflow.keras import regularizers, callbacks
import tensorflow.keras.backend as K 
```

```python
class attention(Layer):
    
    def __init__(self,**kwargs):
        super(attention,self).__init__(**kwargs)
        
    def build(self,input_shape):
        self.W = self.add_weight(name = 'attention_weight', shape = (input_shape[-1],1),
                                 initializer = 'random_normal', trainable = True)
        self.b = self.add_weight(name = 'attention_bias', shape = (input_shape[1],1),
                                 initializer = 'zeros', trainable = True)
        super(attention,self).build(input_shape)
        
    def call(self,x):
        e = K.tanh(K.dot(x,self.W)+self.b)
        e = K.squeeze(e,axis=-1)
        alpha = K.softmax(e)
        alpha = K.expand_dims(alpha,axis=-1)
        context = x*alpha
        context = K.sum(context,axis=1)
        return context
```

```python
input_c1 = Input(shape=[11,11,1])
input_c3 = Input(shape=[11,11,1])
input_c4 = Input(shape=[11,11,1])
input_c5 = Input(shape=[11,11,1])
input_c6 = Input(shape=[11,11,1])
input_c7 = Input(shape=[11,11,1])
input_c9 = Input(shape=[11,11,1])
input_c11 = Input(shape=[11,11,1])

x_c1 = BatchNormalization()(input_c1)
x_c1 = Conv2D(16, 3, activation = 'relu')(x_c1)
x_c1 = MaxPool2D((2,2))(x_c1)
x_c1 = Conv2D(4, 3, activation = 'relu')(x_c1)
flatten_c1 = Flatten()(x_c1)

x_c3 = BatchNormalization()(input_c3)
x_c3 = Conv2D(16, 3, activation = 'relu')(x_c3)
x_c3 = MaxPool2D((2,2))(x_c3)
x_c3 = Conv2D(4, 3, activation = 'relu')(x_c3)
flatten_c3 = Flatten()(x_c3)

x_c4 = BatchNormalization()(input_c4)
x_c4 = Conv2D(16, 3, activation = 'relu')(x_c4)
x_c4 = MaxPool2D((2,2))(x_c4)
x_c4 = Conv2D(4, 3, activation = 'relu')(x_c4)
flatten_c4 = Flatten()(x_c4)

x_c5 = BatchNormalization()(input_c5)
x_c5 = Conv2D(16, 3, activation = 'relu')(x_c5)
x_c5 = MaxPool2D((2,2))(x_c5)
x_c5 = Conv2D(4, 3, activation = 'relu')(x_c5)
flatten_c5 = Flatten()(x_c5)

x_c6 = BatchNormalization()(input_c6)
x_c6 = Conv2D(16, 3, activation = 'relu')(x_c6)
x_c6 = MaxPool2D((2,2))(x_c6)
x_c6 = Conv2D(4, 3, activation = 'relu')(x_c6)
flatten_c6 = Flatten()(x_c6)

x_c7 = BatchNormalization()(input_c7)
x_c7 = Conv2D(16, 3, activation = 'relu')(x_c7)
x_c7 = MaxPool2D((2,2))(x_c7)
x_c7 = Conv2D(4, 3, activation = 'relu')(x_c7)
flatten_c7 = Flatten()(x_c7)

x_c9 = BatchNormalization()(input_c9)
x_c9 = Conv2D(16, 3, activation = 'relu')(x_c9)
x_c9 = MaxPool2D((2,2))(x_c9)
x_c9 = Conv2D(4, 3, activation = 'relu')(x_c9)
flatten_c9 = Flatten()(x_c9)

x_c11 = BatchNormalization()(input_c11)
x_c11 = Conv2D(16, 3, activation = 'relu')(x_c11)
x_c11 = MaxPool2D((2,2))(x_c11)
x_c11 = Conv2D(4, 3, activation = 'relu')(x_c11)
flatten_c11 = Flatten()(x_c11)

concat = concatenate([flatten_c1,flatten_c3,flatten_c4,flatten_c5,flatten_c6,flatten_c7,flatten_c9,flatten_c11])
dense_1 = Dense(32, activation = 'relu', kernel_regularizer = regularizers.l2(0.01))(concat)
dense_2 = Dense(32, activation = 'relu')(dense_1)
dense_3 = Dense(16, activation = 'relu')(dense_2)

output = Dense(1, activation = 'relu')(dense_3)

model = Model(inputs = [input_c1,input_c3,input_c4,input_c5,input_c6,input_c7,input_c9,input_c11],
              outputs = [output]
             )

plot_model(model,show_shapes = True)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0001)
model.compile(optimizer = optimizer,
             loss = 'huber',
             metrics = tf.keras.metrics.RootMeanSquaredError())
callback = callbacks.EarlyStopping(monitor='val_loss',patience=5)

train_output = df_ghi.loc[tr_index,'CSI'].values.reshape(len(tr_index),1)
test_output = df_ghi.loc[te_index,'CSI'].values.reshape(len(te_index),1)
```

```python
history = model.fit([train_c1,train_c3,train_c4,train_c5,train_c6,train_c7,train_c9,train_c11],
          [train_output],
          epochs = 500, batch_size = 64,
          validation_split = 0.2,
          shuffle = True,
          callbacks = callback,
                   )
```

```python
model.evaluate(x = [test_c1,test_c3,test_c4,test_c5,test_c6,test_c7,test_c9,test_c11],
               y = [test_output])
```
